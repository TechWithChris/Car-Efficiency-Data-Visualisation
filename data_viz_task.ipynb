# Data Visualisation 

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
```

```python
# Load data
cars_df= pd.read_csv("Cars93.csv",index_col = 0, keep_default_na=False)
cars_df.head()
```

```python
cars_df.columns
```

### 1. Boxplot

```python
# Boxplot for revs per mile
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


```python
# Plot a box plot showing the revs per mile of each manufacturers Car
plt.figure(figsize= (30,8))
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data= cars_df)
plt.xlabel("Manufacturer")
plt.ylabel("Rev.per.mile")
plt.show()
plt.close
```

- Boxplot answer:The Manufacturer with the highest revs per mile is "Geo"


### 2. Histogram

```python
# Histogram of MPG 
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

```python
# Group data by'Type'
group_type = cars_df.groupby(by=['Type'])

# Get the mean (average) for each type across all columns
cars_df_avg = round(group_type.mean(numeric_only=True), 0) 
cars_df_avg.head()
```

```python
# Create a DataFrame of only the relevant features to plot  
histogram_features = pd.DataFrame({'MPG.highway': cars_df_avg['MPG.highway'],
                                 'MPG.city': cars_df_avg['MPG.city']})
```

```python
# Plot the average miles per gallon (MPG) for highway 
# driving and city driving for each type of car
histogram_features.plot(kind='bar', ylabel="Average MPG")
```

- Histogram answer: It is generally more efficient to drive in the city 

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


```python
# Plot a lineplot of Wheelbase vs Turning circle 
plt.figure()
sns.lineplot(x='Wheelbase', y='Turn.circle', data= cars_df)
plt.title("Wheelbase vs Turning circle")
plt.xlabel("Wheelbase")
plt.ylabel("Turn circle")
plt.show()
plt.close
```

- Lineplot answer: The Turn circle increases as the Wheelbase increaes


### 4. Barplot

```python
# Barplot for mean horsepower
```

```python
# Group data by'Type'
group_type = cars_df.groupby(by=['Type'])

# Get the mean (average) for each type across all columns
# Note the mean can only be calculated for numeric values
cars_df_avg = round(group_type.mean(numeric_only=True), 0) 
cars_df_avg.head()
```

```python
# Create a DataFrame of only the relevant features to plot (Horsepower)
barplot_features= pd.DataFrame({'Horsepower': cars_df_avg['Horsepower']})
```

```python
# Plot a bar chart showing the mean horserpower for the type of cars 
barplot_features.plot.bar(figsize=(5, 5), legend=False, title="Average horsepower for each car type", 
                          ylabel="Average Horsepower", xlabel="Car Type",
                          color='blue')
```

**Barplot question:**  
Does a larger car mean more horsepower?



- Barplot answer:Yes the larger the car the more horsepower
